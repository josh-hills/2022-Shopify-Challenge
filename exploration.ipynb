```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# change display max rows
pd.set_option('display.max_rows', 500)
```

```python
data = pd.read_csv('2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv')
```

```python
data.head(2)
```

```python
#Check if "AOV" is actually the mean
data['order_amount'].mean()
```

```python
sorted = data.sort_values(ascending=False, by='order_amount')
```

```python
# Heres our problem, massive orders
sorted.head(75)
```

```python
plot = sorted.groupby('shop_id').mean()['order_amount'].nlargest(10).plot(kind='bar')
plot.set(xlabel='Store ID', ylabel='AOV', title='Top 10 Stores AOV')
```

```python
sorted['order_amount'].median()
```

```python
# drop all rows where order_amount > 2000
sorted = sorted[sorted['order_amount'] <= 2000]
sorted['order_amount'].mean()
```

```python
sorted.hist(column='order_amount', bins=50)
```
